## Lstm Hyperparameters

```python
lstm_layers=1
hidden_nodes=256
alpha=0.001
keepprob=0.5
batch_size=128
epochs=5
```

```python
vocab_size=len(sorted_split_words)  # total unique words to make embeddings for every word      
# Defining in the default graph
input_x=tf.placeholder(tf.int32,[None,None],name="input")   # placeholders for input,labels and keep prob
label_y=tf.placeholder(tf.int32,[None,None],name="output")
keep_prob=tf.placeholder(tf.float32,name="keep_prob")
```

```python
vocab_size
```

Adding an embedding layer. Instead of one-hot encoding, we build an embedding layer and use that layer as a lookup table.

```python
# Size of embeddings, input layer vecotor dimension for word2vec 
embed_size=300
#embedding=tf.get_variable("word_embeddings",[vocab_size,embed_size])
embedding = tf.Variable(tf.random_uniform((vocab_size+1, embed_size), -1, 1))
embed=tf.nn.embedding_lookup(embedding,input_x)
```

### Lstm Cell

LSTM cells to use in the recurrent network Here we are just defining what the cells look like.

It takes a parameter called num_units, the number of units in the cell, called lstm_size in this code.

Adding dropout to the cell with tf.contrib.rnn.DropoutWrapper. This wraps the cell in another cell, but with dropout added to the inputs and/or outputs.

Here, [drop] * lstm_layers creates a list of cells (drop) that is lstm_layers long. The MultiRNNCell wrapper builds this into multiple layers of RNN cells, one for each cell in the list.

So the final cell in the network is actually multiple (or just one) LSTM cells with dropout.

```python
lstm=tf.contrib.rnn.LSTMCell(hidden_nodes)
drop_out=tf.contrib.rnn.DropoutWrapper(lstm,output_keep_prob=keepprob) #lstm cell wrapper to apply dropout regulariz

#Stacking layers of single lstm cells
stacked_cells=tf.contrib.rnn.MultiRNNCell([drop_out])

#initialising the lstm layers with zero value  (is_tuple can also be used with non zero intial state value)
initial_state=stacked_cells.zero_state(batch_size,tf.float32) 
```

### tf.nn.dynamic_rnn()

```python
#Performs fully dynamic unrolling of inputs.
outputs,final_state= tf.nn.dynamic_rnn(stacked_cells,embed,initial_state=initial_state)
```

**fully_connected** creates a variable called weights, representing a fully connected weight matrix, which is multiplied by the inputs to produce a Tensor of hidden units. If a normalizer_fn is provided (such as batch_norm), it is then applied. Otherwise, if normalizer_fn is None and a biases_initializer is provided then a biases variable would be created and added the hidden units. Finally, if activation_fn is not None, it is applied to the hidden units as well.

**We only care about the final output, we'll be using that as our sentiment prediction. So we need to grab the last output with outputs[:, -1], the calculate the cost from that and labels_.**

```python
predictions=tf.contrib.layers.fully_connected(outputs[:,-1],1,activation_fn=tf.sigmoid)
cost=tf.losses.mean_squared_error(label_y,predictions)
optimizer=tf.train.AdamOptimizer(learning_rate=alpha).minimize(cost)
```

**Calculating predictions and accuracy**

```python
correct_pred=tf.equal(tf.cast(tf.round(predictions),tf.int32),label_y)
accuracy=tf.reduce_mean(tf.cast(correct_pred,tf.float32))
#The functionality of numpy.mean and tensorflow.reduce_mean are the same. They do the same thing
#Computes the mean of elements across dimensions of a tensor.
```

Creating Batches

```python
def get_batches(x,y,batch_size=100):
    n_batch=len(x)//batch_size
    x,y= x[:n_batch*batch_size], y[:n_batch*batch_size]
    for ii in range(0,len(x),batch_size):
        yield x[ii:ii+batch_size],y[ii:ii+batch_size]    
```

### Training 

```python
saver=tf.train.Saver()   # saver object instance

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())    # Initialising every variable
    iteration=1
    for e in range(epochs):
        state=sess.run(initial_state)    # initialising the (zero step of rnn) initial state of rnn cell
        for ii, (x,y) in enumerate(get_batches(train_x,train_y,batch_size),1):
            feed={input_x:x,
                  label_y:y[:,None],
                  keep_prob:keepprob,
                  initial_state:state}       #feed dict for placeholders & initialising for every epoch
            
            loss, state, _ =sess.run([cost,final_state,optimizer],feed_dict=feed)
            
            if iteration%5==0:
                print("Epoch: {}/{}".format(e,epochs),
                      "Iterations: {}".format(iteration),
                      "Training Loss: {:.3f}".format(loss))
                
            if iteration%25==0:
                val_acc=[]     # Checking accuracy on validation set after 25 iters
                val_state=sess.run(stacked_cells.zero_state(batch_size,tf.float32)) # for val set initialising cell
                for x,y in get_batches(val_x,val_y,batch_size):
                    feed={input_x:x,
                          label_y:y[:,None],
                          keep_prob:1,
                          initial_state:val_state}
                    batch_acc,val_state=sess.run([accuracy,final_state],feed_dict=feed)
                    val_acc.append(batch_acc)
                print("Average Validation Accuracy {:.3f}".format(np.mean(val_acc)))
            iteration+=1
            
    saver.save(sess,"/home/pradyuman/lstm/checkpoints.ckpt")
                
```

### Testing the lstm Model

```python
test_acc=[]
with tf.Session() as sess:
    saver.restore(sess,tf.train.latest_checkpoint("/home/pradyuman/lstm/"))
    test_state=sess.run(stacked_cells.zero_state(batch_size,tf.float32))
    for ii , (x,y) in enumerate(get_batches(test_x,test_y,batch_size),1):
        feed={input_x:x,
              label_y:y[:,None],
              keep_prob:1,
              initial_state:test_state}
        batch_acc,test_state=sess.run([accuracy,final_state],feed_dict=feed)
        test_acc.append(batch_acc)
    print("Test Accuracy {:.3f}".format(np.mean(test_acc)))
    
```
